3.Compare and contrast map, filter, and reduce.

Ans :
    
1. map() : The map() function is a type of higher-order. this function takes another function as a parameter along with a sequence of iterables and returns an o/p after applying the function to each iterable present in the sequence.

2. filter() : The filter() function is used to create an o/p list consisting of values for which the function returns true.

3. reduce() : The reduce() function, as the name describes,applies a given function to the iterables and returns a single value.

```python
from functools import reduce
print('Map ->',list(map(lambda x:x+x, [1,2,3,4])))
print('Filter ->' , list(filter(lambda x:x%2 !=0,[1,2,3,4])))
print('Reduce ->' ,reduce(lambda x,y:x+y,[1,2,3,4,5,6]))
```
